# Connect to Google Drive

```python
from google.colab import drive
drive.mount('/content/drive')
```

# Import and initialize PySpark session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("TBC_TRANSACTIONS") \
    .getOrCreate()
```

# Read TBC transactions CSV file

```python
file_path = "/content/drive/MyDrive/Statement_1.csv"

transactions_raw = spark.read.csv(file_path, header=True, inferSchema=True)
```

# Display unique operation types and find the one we're insterested in: incomes

```python
from pyspark.sql.functions import col

transactions_raw.select(col("`Op. Code`")).distinct()
```

## Seems like the transactions with "Op. Code"=='MMS' are the ones we're looking for

```python
df = transactions_raw.filter(col("`Op. Code`") == 'MMS')
```

# Write all the filtered transactions back to the Google Drive as new CSV file

```python
df.write.format("csv") \
    .option("header", "true") \
    .save("/content/drive/MyDrive/TBC_Statements/TBC_Transactions.csv")
```
